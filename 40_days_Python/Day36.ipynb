```python
import sqlite3
import pandas as pd

connection = sqlite3.connect('day36_db.sqlite')
cursor = connection.cursor()
```

```python
# Create table
cursor.execute('''
CREATE TABLE STOCK(
  ID INTEGER PRIMARY KEY,
  NAME TEXT NOT NULL,
  BUY INTEGER NOT NULL,
  INVESTOR TEXT NOT NULL);
''')

# Insert values
cursor.execute('''
INSERT INTO STOCK(ID,NAME,BUY,INVESTOR)
VALUES
  (1, 'ACB', 29.45, 'Nguyen'),
  (2, 'VIC', 44.55, 'Nguyen'),
  (3, 'GMD', 74.30, 'Nguyen'),
  (4, 'ACB', 28.45, 'Vinh'),
  (5, 'VIC', 40.55, 'Vinh'),
  (6, 'GMD', 60.30, 'Vinh');
''')

connection.commit()

data = pd.read_sql_query('SELECT * FROM STOCK',connection)
print(data)
```

```python
# Ex 1:
query = '''
SELECT SUM(BUY) AS Tong_gia_ban
FROM STOCK;
'''

data_sum = pd.read_sql_query(query,connection)
print(data_sum)
```

```python
# Ex 2:
query = '''
SELECT INVESTOR, NAME, MAX(BUY) AS MAX_PRICE
FROM STOCK
GROUP BY INVESTOR;
'''

data_max_by_investor = pd.read_sql_query(query,connection)
print(data_max_by_investor)
```
