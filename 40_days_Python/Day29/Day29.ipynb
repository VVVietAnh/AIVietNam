```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = librosa.example('nutcracker')

audio, sr = librosa.load(filename)

time = np . linspace (0 , audio . shape [0] / sr , num = audio . shape [0])
plt . plot ( time , audio , color ="blue")
plt . xlabel (" Time (s)")
plt . ylabel (" Amplitude ( quantized )")
plt . title (" Wav file visualization ")
plt . axhline ( y =0 , color ='r', linestyle ='-')
plt . show ()
```

```python
# create file.wav
import soundfile as sf
sf.write('file.wav', audio, sr)
```

### speech_recognition

```python
# !pip install SpeechRecognition
```

```python
import speech_recognition as sr

r = sr.Recognizer()
audio_filename = 'file.wav'

my_audio = sr.AudioFile(audio_filename)
with my_audio as source:
    audio = r.record(source)

print(type(audio))
your_speech = r.recognize_google(audio, language='en-US')
print("Audio transcription: ", your_speech)
```

### gtts

```python
# !pip install gtts
```

```python
from gtts import gTTS
import librosa
import numpy as np

lang = 'vi'
output_filename = 'record.mp3'
content = "xin chào mọi người"
output = gTTS(content, lang=lang, slow=False)   # text to speech

output.save(output_filename)    # save google audio to a file
data, sr = librosa.load(output_filename)    # load google audio file

import IPython.display as ipd
ipd.display(ipd.Audio(data, rate=sr))    # play google audio file
```

```python
# Ex1 - show the graph and transcribe the audio file
test_case_1 = "iloveyou.wav"
data, sr = librosa.load(test_case_1)
time = np.linspace(0, data.shape[0] / sr, num=data.shape[0])
plt.plot(time, data, color="blue")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (quantized)")
plt.title("Wav file visualization")
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
import speech_recognition as sr

r = sr.Recognizer()
my_audio = sr.AudioFile(test_case_1)
with my_audio as source:
    audio = r.record(source)

your_speech = r.recognize_google(audio, language='vi-VN')
print("Audio transcription: ", your_speech)
```
