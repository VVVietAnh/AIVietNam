```python
from underthesea import pos_tag, sent_tokenize, sentiment, word_tokenize, classify

text = 'Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hất chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'

print(f"POS Tagging: {pos_tag(text)}")
print(f"Text Classification: {classify(text)}")
print(f"Sentiment Analysis: {sentiment(text)}")
print(f"Sentence Segmentation: {sent_tokenize(text)}")
print(f"Word Segmentation: {word_tokenize(text)}")
```
