```python
from translate import Translator

translator = Translator(to_lang="vi")
translation = translator.translate("Don't cry because it's over, smile because it happend.")
print(translation)
```

```python
from googletrans import Translator

translate = Translator()
result = translate.translate('Hôm nay thời tiết không tốt và mưa nhiều', dest='fr')

print(result)
print(result.text)
```

```python
from googletrans import Translator

translator = Translator()

text = 'Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hất chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'

eng_translate = translator.translate(text, dest='en')
jp_translate = translator.translate(text, dest='ja')

print(eng_translate.text)
print(jp_translate.text)
```
